```python
## QUES 5)  Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
# heatmap.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# If your 'data' DataFrame doesn't have numerical features for correlation,
# you'll need to create or load a dataset with such features.
# Here's an example of creating a synthetic dataset:
num_features = 3  # Number of numerical features
num_samples = 100

# Create random data with some correlation
data_num = pd.DataFrame(np.random.rand(num_samples, num_features), columns=[f'Feature_{i}' for i in range(num_features)])

# Introduce some correlation between features
data_num['Feature_1'] = data_num['Feature_0'] * 2 + np.random.rand(num_samples)
data_num['Feature_2'] = data_num['Feature_1'] - data_num['Feature_0'] + np.random.rand(num_samples)

# Now, 'data_num' contains numerical features for correlation analysis.
# Calculate the correlation matrix
corr_matrix = data_num.corr() # Changed to data_num to use numerical features

# Create a heatmap
sns.set_theme(style="white")
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, linewidths=0.5, linecolor='black')

# Add title
plt.title('Correlation Matrix Heatmap')

# Show plot
plt.show()
```

```python
## QUES 1) Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
import numpy as np
from bokeh.plotting import figure, show, output_file

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)
# Calculate y values as the sine of x
y = np.sin(x)

# Configure the output to display in a separate HTML file
output_file("sine_wave.html")

# Create a new figure
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='y')

# Add a line plot for the sine wave
p.line(x, y, legend_label="y=sin(x)", line_width=2)

# Show the results
show(p)
```

```python
## QUES 2) Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
# markers based on the 'sizes' and 'colors' columns.
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource

# Set the random seed for reproducibility
np.random.seed(42)

# Generate random data
data = {
    'x': np.random.rand(100) * 100,
    'y': np.random.rand(100) * 100,
    'sizes': np.random.rand(100) * 50 + 10,  # Sizes between 10 and 60
    'colors': np.random.rand(100) * 100      # Colors between 0 and 100
}

# Create a DataFrame
df = pd.DataFrame(data)

# Configure the output to display in the notebook
output_notebook()

# Create a ColumnDataSource
source = ColumnDataSource(df)

# Create a new figure
p = figure(title="Bokeh Scatter Plot with Different Sizes and Colors",
           x_axis_label='X', y_axis_label='Y')

# Add a scatter plot to the figure
p.scatter('x', 'y', size='sizes', color='colors', alpha=0.6, source=source)

# Show the results
show(p)
```

```python
## QUES 4) Create a Bokeh histogram to visualize the distribution of the given data.
"""data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)"""
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook
import numpy as np # Import numpy for random data generation and histogram calculation


# Configure the output to display in the notebook
output_notebook()

# Generate random data and calculate histogram
data_hist = np.random.randn(1000)  # Generate 1000 random data points
hist, edges = np.histogram(data_hist, bins=30)  # Calculate histogram data

# Create a new plot with a title and axis labels
p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency')

# Add a quad glyph to create the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.7)

# Show the results
show(p)
```

```python
## QUES 5)  Create a Bokeh heatmap using the provided dataset
"""data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)"""
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.transform import transform
from bokeh.palettes import Viridis256
import numpy as np # Import numpy for data generation

# Configure the output to display in the notebook
output_notebook()

# Generate data for the heatmap - This was previously commented out
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create a new plot with a title and axis labels
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y',
           x_range=(0, 1), y_range=(0, 1),
           toolbar_location=None, tools="")

# Create a color mapper
color_mapper = LinearColorMapper(palette=Viridis256, low=data_heatmap.min(), high=data_heatmap.max())

# Add the heatmap to the plot
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

# Add a color bar
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the results
show(p)
```
